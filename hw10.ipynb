### Problem 3
Consider $x = -t^2/4$.  Then $x'=-t/2$.  Furthermore,
\begin{align*}
\sqrt{t^2 + 4(-t^2/4)} - t &= \sqrt{t^2 - t^2} - t \\
&= -t \\
&= 2x'
\end{align*}
and $x(2) = -2/2 = -1$.

Now consider $x = 1 - t$.  Then $x' = 1$ and
\begin{align*}
\sqrt{t^2 + 4(1 - t)} - t &= \sqrt{t^2 - 4t + 4 } - t \\
&= \sqrt{(t - 2)^2} - t \\
&= |t - 2| - t \\
&= -2\text{ if }t\geq 2\\
&= 2x'
\end{align*}
and $x(2) = -1$.

In this case, we have $f(t,x) = \sqrt{t^2 + 4x} - t$, which is continuous only when $t^2 + 4x > 0$.  In particular, $(2, -1)$ lies on the boundary of this region since $2^2 + 4(-1) = 0$.  For any neighborhood of $(2, -1)$, we can find a point where $f$ is undefined, e.g., for any $\varepsilon > 0$, $f$ is undefined at $(2, -1 - \varepsilon)$.  So this example does not contradict the Picard-Lindelof theorem.

### Problem 13
Consider the initial value problem $x' = \sqrt{|x|}$, $x(0) = 0$.  First note that the zero function $x = 0$ is a solution.

Note also that this is a separable equation, so we can solve by rewriting as
\begin{align*}
\frac{1}{\sqrt{|x|}} x' &= 1 \\
|x|^{-1/2} x' &= 1 \\
\int |x|^{-1/2}\,dx &= \int \,dt  \\
\end{align*}

At this point, let's assume $x\geq 0$, so that $|x| = x$.  This gives us
\begin{align*}
\int x^{-1/2}\,dx &= t + C \\
2x^{1/2} &= t + C \\
\sqrt x &= \frac{t + C}{2} \\
x &= \left(\frac{t + C}{2}\right)^2 \\
\end{align*}

Using the initial condition $x(0) = 0$,  we have
\begin{align*}
0 &= \left(\frac{C}{2}\right)^2 \\
C & = 0
\end{align*}

So we see that $x = t^2/4$ is another solution.

If $f(t,x) = \sqrt{|x|}$, then $f$ is continuous everywhere.  However,
\begin{equation*}
\frac{\partial f}{\partial x} = \begin{cases}
\frac{1}{2\sqrt x} &\text{if } x > 0 \\
-\frac{1}{2\sqrt{-x}} &\text{if }x < 0,
\end{cases}
\end{equation*}
which is undefined at 0.  Therefore, it does not satisfy the conditions of the Picard-Lindelof theorem to guarantee uniqueness.


## Section 8.2
### Rewrite $x'''$ in terms of $f$ and its partial derivatives

We have already seen that $x''(t) = \frac{\partial f}{\partial t}(t, x(t)) + \frac{\partial f}{\partial x}(t,x(t)) f(t,x(t))$  In other words, $x'' = (g\circ\mathbf{r})(t)$, where $g(t,x) = \frac{\partial f}{\partial t}(t, x) + \frac{\partial f}{\partial x}(t,x) f(t,x)$ and $\mathbf r(t) = \langle t, x(t)\rangle$.

So by the multivariable chain rule, we see that $x'''(t) = \nabla g(t,x(t))\cdot\mathbf r'(t)$.

We have
\begin{align*}
\frac{\partial g}{\partial t} &= \frac{\partial}{\partial t}\left(\frac{\partial f}{\partial t} + \frac{\partial f}{\partial x}f\right) \\
&= \frac{\partial^2 f}{\partial t^2} + \frac{\partial^2 f}{\partial t\partial x}f + \frac{\partial f}{\partial x}\frac{\partial f}{\partial t} \\
\frac{\partial g}{\partial x} &= \frac{\partial}{\partial x}\left(\frac{\partial f}{\partial t} + \frac{\partial f}{\partial x}f\right) \\
&= \frac{\partial^2 f}{\partial x\partial t} + \frac{\partial^2 f}{\partial x^2}f + \left(\frac{\partial f}{\partial x}\right)^2 \\
\mathbf r'(t) &= \langle 1, x'\rangle \\
&= \langle 1, f\rangle
\end{align*}

Putting this together,
\begin{align*}
x'''(t) &= \left\langle\frac{\partial g}{\partial t}, \frac{\partial g}{\partial x}\right\rangle\cdot\langle 1, f\rangle \\
&= \frac{\partial^2 f}{\partial t^2} + \frac{\partial^2 f}{\partial t\partial x}f + \frac{\partial f}{\partial x}\frac{\partial f}{\partial t} + \left( \frac{\partial^2 f}{\partial x\partial t} + \frac{\partial^2 f}{\partial x^2}f + \left(\frac{\partial f}{\partial x}\right)^2\right)f \\
&= \frac{\partial^2 f}{\partial t^2} + \frac{\partial f}{\partial x}\frac{\partial f}{\partial t} + \left(2\frac{\partial^2 f}{\partial x\partial t} + \left(\frac{\partial f}{\partial x}\right)^2\right) f + \frac{\partial^2 f}{\partial x^2}f^2
\end{align*}

### Macaulay2 code for third order Taylor method

Let's adapt our result into a Macaulay2 function.  

```macaulay2
taylorMethod3 = (f, dfdt, dfdx, d2fdt2, d2fdtdx, d2fdx2, t, x, h, n) -> (
    for i from 1 to n list (
	x' := f(t, x);
	x'' := dfdt(t, x) + dfdx(t,x) * x';
	x''' := d2fdt2(t,x) + dfdx(t,x) * dfdt(t,x) + (
	    2*d2fdtdx(t,x) + (dfdx(t,x))^2) * x' + d2fdx2(t,x) * x'^2;
	x += h * (x' + h/2 * (x'' + h/3 * x'''));
	t += h;
	(t, x)))
```

# Computer Problem 3

In this case, we have
\begin{align*}
f(t,x) &= e^{-t^2} \\
\frac{\partial f}{\partial t} &= -2te^{-t^2} \\
\frac{\partial^2 f}{\partial t^2} &= -2t\cdot(-2te^{-t^2}) - 2e^{-t^2} \\
&= (4t^2 - 2)e^{-t^2}\\
\frac{\partial f}{\partial x} = \frac{\partial^2 f}{\partial t\partial x} = \frac{\partial^2 f}{\partial x^2} &= 0
\end{align*}

```macaulay2
printWidth = 0 -- so our output isn't super long
printingPrecision = 0
f = (t,x) -> exp(-t^2)
dfdt = (t,x) -> -2*t*exp(-t^2)
d2fdt2 = (t,x) -> (4*t^2 - 2)*exp(-t^2)
zerof = (t,x) -> 0
taylorMethod3(f, dfdt, zerof, d2fdt2, zerof, zerof, 0, 0, 0.01, 200)
```

Our last estimate is quite close to the value of $\operatorname{erf} 2$ given in the textbook!

### Problem 7
We have
\begin{align*}
f(t,x) &= \sin(t^2) \\
\frac{\partial f}{\partial t} &= 2t\cos(t^2) \\
\frac{\partial^2 f}{\partial t^2} &= 2t\cdot(-2t\sin(t^2)) + 2\cos(t^2) \\
&= 2\cos(t^2) - 4t^2\sin(t^2)
\end{align*}
As in the previous problem, the remaining derivatives are all zero.

```macaulay2
f = (t,x) -> sin(t^2)
dfdt = (t,x) -> 2*t*cos(t^2)
d2fdt2 = (t,x) -> 2*cos(t^2) - 4*t^2*sin(t^2)
taylorMethod3(f, dfdt, zerof, d2fdt2, zerof, zerof, 0, 0, 0.1, 100)
```

Let's compute the corresponding values of the Fresnel sine function using Macaulay2's Python interface to compare.

```macaulay2
needsPackage "Python"
scipy = import "scipy"
apply(100, i -> (
    t := 0.1 * (i + 1);
    (t, (scipy@@special@@fresnel t)_0)))
```

Here's a [Desmos graph](https://www.desmos.com/calculator/nwy4o0vpdn) of the two data sets.  Our estimated solution has the same basic behavior, but it diverges pretty quickly from the actual value.  Perhaps this is why Kincaid and Cheney said to use the 5th order Taylor method!

### Problem 12
Here we have
\begin{align*}
f(t,x) &= \sqrt{1 + t^3} \\
\frac{\partial f}{\partial t} &= \frac{3t^2}{2\sqrt{1 + t^3}} \\
\frac{\partial^2 f}{\partial t^2} &= \frac{2\sqrt{1+t^3}\cdot 6t - 3t^2\cdot\frac{3t^2}{\sqrt{1+t^3}}}{4(1+t^3)} \\
&= \frac{12t(1+t^3)-9t^4}{4(1+t^3)\sqrt{1+t^3}} \\
&= \frac{3t^4 + 12t}{4(1+t^3)\sqrt{1+t^3}} \\
\end{align*}
And as in the previous problems, the remaining derivatives are all 0.

```macaulay2
f = (t,x) -> sqrt(1 + t^3)
dfdt = (t,x) -> 3*t^2/(2*sqrt(1 + t^3))
df2dt2 = (t,x) -> (3*t^4 + 12*t)/(4*(1+t^3)*sqrt(1+t^3))
taylorMethod3(f, dfdt, zerof, d2fdt2, zerof, zerof, 0, 0, numeric(1/64), 5*64)
```

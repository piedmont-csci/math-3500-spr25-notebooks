#### Part (a)

Suppose $x = \frac{3}{5}$.  We would like to find the sequence $(a_n)$ for which $x = (0.a_1a_2\cdots)_2$.  

Multiplying by 2:

$$\frac{6}{5} = (a_1.a_2a_3\cdots)_2 \implies a_1 = 1\text{ since }\frac{6}{5}\geq 1$$

Subtracting 1:

$$\frac{1}{5} = (0.a_2a_3\cdots)_2$$

Multiplying by 2:

$$\frac{2}{5} = (a_2.a_3a_4\cdots)_2\implies a_2 = 0\text{ since }\frac{2}{5}< 1$$

Multiplying by 2:

$$\frac{4}{5} = (a_3.a_4a_5\cdots)_2\implies a_3 = 0\text{ since }\frac{4}{5} < 1$$

Multiplying by 2:

$$\frac{8}{5} = (a_4.a_5a_6\cdots)_2\implies a_4 = 1\text{ since }\frac{8}{5}\geq 1$$

Subtracting 1:

$$\frac{3}{5} = (0.a_5a_6\cdots)_2$$

Now we're back where we started, and in particular, we see that $a_{n+4} = a_n$ for all $n$.  Consequently:

\begin{align*}
\frac{3}{5} &= (0.1001100110011001\cdots)_2 \\
&= (1.001100110011001\cdots)_2\times 2^{-1}
\end{align*}

Now let's find the closest machine words by rounding the mantissa to the nearest $2^{-23}$ place in each direction:

\begin{align*}
x_- &= (1.00110011001100110011001)_2\times 2^{-1} \\
x_+ &= (1.00110011001100110011010)_2\times 2^{-1}
\end{align*}

Note that
\begin{align*}
x - x_- &= (0.00000000000000000000000100110011001\cdots)\times 2^{-1} \\
&= (1.00110011001\cdots)\times 2^{-24}\times 2^{-1} \\
&= \frac{3}{5}\times 2^{-24}
\end{align*}
and
\begin{align*}
x_+ - x &= (x_+ - x_-) - (x - x_-) \\
&= 2^{-23}\times 2^{-1} - \frac{3}{5}\times 2^{-24} \\
&= 1\times 2^{-24} - \frac{3}{5}\times 2^{-24} \\
&= \frac{2}{5}\times 2^{-24}
\end{align*}

Since $x_+$ has the smaller error, we use $\operatorname{fl}(x) = x_+$, and the relative roundoff error is

\begin{align*}
\frac{x_+ - x}{x} &= \frac{\frac{2}{5}\times 2^{-24}}{\frac{3}{5}} \\
&= \frac{2}{3}\times 2^{-24}
\end{align*}

\begin{align*}
\frac{x - x_-}{x} &\leq \frac {2^{-24}}{1 + 2^{-24}} \\

### Problem 7
Recall that a normalized single-precision machine number is of the form $(-1)^sq\times 2^m$, where $s\in\{0,1\}$, $q = (1.a_1\cdots a_{23})_2$, and $-126\leq m\leq 127$.

So we have 2 choices for $s$, $2^{23}$ choices for the bits of $q$, and $127 - (-126) + 1 = 254$ choices for $m$.  Consequently, there are
$$2 \cdot 2^{23} \cdot 254 = 4,261,412,864$$
normalized single-precision machine numbers.

### Problem 8
Suppose a normalized machine number has two representations, i.e., there exist integers $s,q,m,t,r,n$ such that
$$(-1)^sq\times 2^m = (-1)^t r\times 2^n$$
where $1\leq q,r < 2$.

Certainly, $s = t$ since the number is either positive or negative.  So $q\times 2^m = r\times 2^n$.

Suppose $m = n$.  Then, dividing both sides by $2^m$, we see that $q = r$ and we are done.

Now suppose $m\neq n$.  Without loss of generality, we may assume that $m > n$, or equivalently, $m - n \geq 1$.

Then, dividing by $2^n$, we get:
$$r = q\times 2^{m - n} \geq q\times 2^1$$

But since $q\geq 1$, this means that $r\geq 2$.  But we know that $q < 2$, a contradiction.

Consequently, $s = t$, $m = n$, and $q = r$, i.e., normalized machine numbers have unique representations.

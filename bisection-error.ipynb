# Proof of bisection method error bound
## Kincaid & Cheney Section 3.1 Theorem 1

**Theorem** If $(a_n)$ is the sequence of left endpoints, $(b_n)$ the sequence of right endpoints, and $(c_n)$ the sequence of midpoints in the bisection method, and if $a_n\leq r\leq b_n$ for all $n$, then
$$|r - c_n|\leq 2^{-(n+1)}|b_0 - a_0|$$

*Proof.*  First, we make an important observation.  For all $n$,

\begin{align*}
r - c_n &= r - \frac{a_n + b_n}{2} \\
\end{align*}

Since $r\geq a_n$ for all $n$, we have

\begin{align*}
r - c_n &\geq a_n - \frac{a_n + b_n}{2}\\
&= \frac{2a_n - a_n - b_n}{2}\\
&= - \frac{b_n - a_n}{2}
\end{align*}

Since $r\leq b_n$ for all $n$, we have
\begin{align*}
r - c_n &\leq b_n - \frac{a_n + b_n}{2}\\
&= \frac{2b_n - a_n - b_n}{2}\\
&= \frac{b_n - a_n}{2}
\end{align*}

Consequently,
$$-\frac{b_n - a_n}{2} \leq r - c_n \leq \frac{b_n - a_n}{2}$$

Or equivalently,
$$|r - c_n| \leq \frac{b_n - a_n}{2}$$

So it is sufficient to prove the following result:

$$\frac{b_n - a_n}{2} = 2^{-(n + 1)}(b_0 - a_0)$$
for all $n$.

Now we proceed by induction on $n$.

*Base case*.  When $n = 0$, we have
$$\frac{b_0 - a_0}{2} = -2^{-(0 + 1)}(b_0 - a_0)$$

*Inductive hypothesis*.  Suppose for some $k\geq 0$,
$$\frac{b_k - a_k}{2} = 2^{-(k + 1)}(b_0 - a_0)$$

**Case 1.** Suppose $a_{k+1} = a_k$ and $b_{k+1} = c_k = \frac{a_k + b_k}{2}$.

Then
\begin{align*}
\frac{b_{k+1} - a_{k+1}}{2} &= \frac{\frac{a_k + b_k}{2} - a_k}{2} \\
&= \frac{a_k + b_k - 2a_k}{4} \\
&= \frac{1}{2}\cdot\frac{b_k - a_k}{2}\\
&= \frac{1}{2}\cdot 2^{-(k+1)}(b_0 - a_0) \\
&= 2^{-(k+2)}(b_0 - a_0) \\
\end{align*}

**Case 2.** Suppose $a_{k+1} = c_k = \frac{a_k + b_k}{2}$ and $b_{k+1} = b_k$.

Then
\begin{align*}
\frac{b_{k+1} - a_{k+1}}{2} &= \frac{\frac{b_k - a_k}{2} + b_k}{2} \\
&= \frac{b_k - a_k + 2b_k}{4} \\
&= \frac{1}{2}\cdot\frac{b_k - a_k}{2}\\
&= \frac{1}{2}\cdot 2^{-(k+1)}(b_0 - a_0) \\
&= 2^{-(k+2)}(b_0 - a_0) \\
\end{align*}

We have now proved by induction that 
$$\frac{b_n - a_n}{2} = 2^{-(n + 1)}(b_0 - a_0)$$,

and since we have proved that $|r - c_n|\leq \frac{b_n - a_n}{2}$, it follows that
$$|r - c_n|\leq 2^{-(n+1)}|b_0 - a_0|$$
for all $n$.
